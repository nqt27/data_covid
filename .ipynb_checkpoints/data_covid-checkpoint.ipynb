```python
# Import thư viện
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu từ DataFrame
df = pd.read_csv('data.csv')

# Hiển thị 5 dòng đầu tiên của dữ liệu
df.head()
```

```python
print("Số lượng dòng và cột: ", df.shape)
print("Tổng số lượng dữ liệu bị thiếu: ", df.isnull().sum().sum())
print("Thông tin về kiểu dữ liệu của từng cột:\n", df.dtypes)
```

## Các biểu đồ phân tích

```python
# Biểu đồ đám mây so sánh số ca nhiễm và số ca tử vong theo quốc gia
plt.figure(figsize=(10, 8))
plt.scatter(df['Confirmed'], df['Deaths'], alpha=0.7, edgecolors='w', linewidth=0.5)
plt.title('Biểu Đồ Đám Mây: Số Ca Nhiễm vs. Số Ca Tử Vong')
plt.xlabel('Số Ca Nhiễm')
plt.ylabel('Số Ca Tử Vong')
plt.grid(True)
plt.show()
```

```python
# Chọn Top 10 quốc gia theo số ca nhiễm
top_countries = df.sort_values(by='Confirmed', ascending=False).head(10)

# Biểu đồ cột
plt.figure(figsize=(12, 8))
sns.barplot(x='Confirmed', y='Country_Region', data=top_countries)
plt.title('Top 10 Quốc Gia với Số Ca Nhiễm Cao Nhất')
plt.xlabel('Số Ca Nhiễm')
plt.ylabel('Quốc Gia')
plt.show()
```

```python
# Tính số ca nhiễm mới hàng ngày
df['NewCases'] = df['Confirmed'].diff()

# Biểu đồ đường
plt.figure(figsize=(12, 6))
sns.lineplot(x='Last_Update', y='NewCases', data=df, marker='o')
plt.title('Biểu Đồ Đường: Số Ca Nhiễm Mới Theo Ngày')
plt.xlabel('Ngày')
plt.ylabel('Số Ca Nhiễm Mới')
plt.show()
```

```python
# Chuyển cột 'Last_Update' sang định dạng datetime
df['Last_Update'] = pd.to_datetime(df['Last_Update'])

# Biểu đồ đường thể hiện sự biến động theo thời gian
plt.figure(figsize=(12, 6))
sns.lineplot(x='Last_Update', y='Confirmed', data=df, marker='o')
plt.title('Biểu Đồ Đường: Số Ca Nhiễm Theo Thời Gian')
plt.xlabel('Ngày')
plt.ylabel('Số Ca Nhiễm')
plt.show()
```

```python
# Chọn một số quốc gia để so sánh
selected_countries = ['Vietnam', 'US', 'India', 'Brazil', 'Russia']

# Lọc dữ liệu cho các quốc gia được chọn
selected_df = df[df['Country_Region'].isin(selected_countries)]

# Tính tổng số ca nhiễm cho mỗi quốc gia
total_cases_by_country = selected_df.groupby('Country_Region')['Confirmed'].max()

# Biểu đồ tròn
plt.figure(figsize=(10, 10))
plt.pie(total_cases_by_country, labels=total_cases_by_country.index, autopct='%1.1f%%', startangle=90)
plt.title('Biểu Đồ Tròn: Tỉ Lệ Số Ca Nhiễm Theo Quốc Gia được Chọn')
plt.show()
```
